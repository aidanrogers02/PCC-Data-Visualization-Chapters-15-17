Random Walks (series of random decisions reflected in the real world):

Random Walk Class

```python
#@title Random Walk Class
from random import choice

class RandomWalk:
  """A class to generate random walks."""
  
  def __init__(self, num_points=5000):
    """Initialize attributes of a walk."""
    self.num_points = num_points
    # All walks start at (0,0).
    self.x_values = [0]
    self.y_values = [0]

  def get_step(self):
    """ Determine the distance and direction of a step"""
    direction = choice([1, -1])
    distance = choice([0, 1, 2, 3, 4])
    step = direction * distance

    return step


  def fill_walk(self):
    """Calculate all the points in the walk."""
    # Keep taking steps until the walk reaches the desired length"""
    while len(self.x_values) < self.num_points:
      # Decide which direction to go and how far to go in that direction
      x_step = self.get_step()

      y_step = self.get_step()

      # Reject moves that go nowhere
      if x_step == 0 and y_step == 0:
        continue 
      
      # Calculate the new position
      x = self.x_values[-1] + x_step
      y = self.y_values[-1] + y_step

      self.x_values.append(x)
      self.y_values.append(y)
```

Rolling Dice with Plotly:

Dice Class

```python
#@title Dice Class
from random import randint

class Die:
  """A class representing a single die"""
  def __init__(self, num_sides=6):
    """Assume a six-sided die"""
    self.num_sides = num_sides

  def roll(self):
    """Return a random value between 1 and a number of sides"""
    return randint(1, self.num_sides)
```

Making a Histogram with a Die

```python
#@title Making a Histogram with a Die
from plotly.graph_objs import Bar, Layout
import plotly.graph_objects as go
from plotly import offline

# create a D6
die = Die()

# make some rolls, and store the results in a list
results = []
for roll_num in range(1000):
  result = die.roll()
  results.append(result)


# Analyze results
frequencies = []
for value in range(1, die.num_sides+1):
  frequency = results.count(value)
  frequencies.append(frequency)

# Visualize the results

# convert to a list since Bar needs a list
x_values = list(range(1, die.num_sides+1))
# [Bar(list of x values, list of y values)]
data = [Bar(x=x_values, y=frequencies)]

x_axis_config = {'title': 'Result'}
y_axis_config = {'title': 'Frequency of Result'}
my_layout = Layout(title='Results of rolling one D6 1000 times', 
                   xaxis=x_axis_config, yaxis=y_axis_config)
# Generates the Bar graph, needs dictionary with the data and layout, and
# filename to save to
offline.plot({'data': data, 'layout': my_layout}, filename='d6.html')

fig = go.Figure({'data': data, 'layout': my_layout})
fig.show()

```

Plot the high and low temperatures over a year



```python
#@title Plot the high and low temperatures over a year
import csv
from datetime import datetime
import matplotlib.pyplot as plt

filename = 'drive/MyDrive/csv_file_data/death_valley_2018_simple.csv'
with open(filename) as f:
  # reads in the file in the proper format for csv files
  reader = csv.reader(f)
  # next(csvfile) will grab an entire line at a time, the reader
  # will now be a line further down
  header_row = next(reader)
  

  # # enumerate returns the index of each item and the value of each item
  for index, column_header, in enumerate(header_row):
  #   print(index, column_header)
    if column_header == 'TMAX':
      highin = index
    if column_header == "TMIN":
      minin = index




  # Get dates, and high and low temperatures from this file
  dates, highs, lows = [], [], []
  for row in reader:
    # datetime.strptime can format date in a readable way
    current_date = datetime.strptime(row[2], '%Y-%m-%d')
    # try-except block incase data is missing
    try: 
      # every fourth index in a row is the high temp
      high = int(row[highin])
      low = int(row[minin])
    except ValueError:
      print(f"Missing data for {current_date}")
    else:
      highs.append(high)
      lows.append(low)
      dates.append(current_date)
  
  # Plot the high and temperatures
  plt.style.use('seaborn')
  fig, ax = plt.subplots()
  # alpha changes the opacity
  ax.plot(dates, highs, c='red', alpha=0.5)
  ax.plot(dates, lows, c='blue', alpha=0.5)
  # .fill_between(x, y1, y2) will take a series of x values and two series of 
  # y values and fill the space between the y valeus
  ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.1)

  # Format plot
  ax.set_title("Daily high and low temperatures - 2018\nDeath Valley, CA", fontsize=20)
  ax.set_xlabel('', fontsize=16)
  # Draws date labels diagonally to make sure they don't overlap
  fig.autofmt_xdate()
  ax.set_ylabel("Temperature (F)", fontsize=16)
  ax.tick_params(axis='both', which='major', labelsize=16)

  plt.show()
```


Compare temperatures between Sitka, AL and Death Valley, CA

```python
#@title Compare temperatures between Sitka, AL and Death Valley, CA
import csv
from datetime import datetime
import matplotlib.pyplot as plt

def create_data(filename, dates, highs, lows):
  """Get info on temperature needed to graph"""
  # Open file
  with open(filename) as f:
    reader = csv.reader(f)
    header_row = next(reader)


    # Find indexes for what column high and low represent
    high_in, low_in, date_in = 0, 0, 0
    for index, column_header in enumerate(header_row):
      if(column_header == "TMAX"):
        high_in = index
      if(column_header == "TMIN"):
        low_in = index
      if(column_header == "DATE"):
        date_in = index


    # Read through file row by row inputting data into appropriate lists
    for row in reader:
      current_date = datetime.strptime(row[date_in], '%Y-%m-%d')
      try:
        high = int(row[high_in])
        low = int(row[low_in])
      except ValueError:
        continue
      else:
        lows.append(low)
        highs.append(high)
        dates.append(current_date)
      

filename1 = 'drive/MyDrive/csv_file_data/sitka_weather_2018_simple.csv'
filename2 = 'drive/MyDrive/csv_file_data/death_valley_2018_simple.csv'

# Create lists to hold graph info
dates, highs, lows = [], [], []

# Fill data for Stika, AK
create_data(filename1, dates, highs, lows)


plt.style.use('seaborn')
fig, ax = plt.subplots()

ax.plot(dates, highs, c='red', alpha=0.7)
ax.plot(dates, lows, c='blue', alpha=0.7)

ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.3)

# Create data for Death Valley, CA
dates, highs, lows = [], [], []
create_data(filename2, dates, highs, lows)

ax.plot(dates, highs, c='red', alpha=0.4)
ax.plot(dates, lows, c='blue', alpha=0.4)

ax.fill_between(dates, highs, lows, facecolor='blue', alpha=0.2)

ax.set_title("Average Daily Temperature in Death Valley, CA vs Sitka, AK", fontsize=20)
ax.set_xlabel('', fontsize=16)

fig.autofmt_xdate()
ax.set_ylabel("Temperature (F)", fontsize=16)
ax.tick_params(axis='both', which='major', labelsize=16)

plt.show()
```

Mapping Global Data Sets: JSON Format

Creating a Plotly of Earthquakes Over a 30 Day Period

```python
#@title Creating a Plotly of Earthquakes Over a 30 Day Period
import json

from plotly.graph_objs import Scattergeo, Layout
import plotly.graph_objects as go
from plotly import offline

# Explore the stucture of the data.
filename = 'drive/MyDrive/csv_file_data/eq_data_30_day_m1.json'
with open(filename) as f:
  # json.load() converts the data into a format Python can work with
  all_eq_data = json.load(f)

readable_file = 'drive/MyDrive/csv_file_data/readable_eq_data.json'
with open(readable_file, 'w') as f:
  # json.dump() takes a JSON data object and a file object, and writes the data
  # to that file. indent = 4 tells dump() to format the data using indentation
  # that mathces the data's structure
  json.dump(all_eq_data, f, indent=4)

# Make a list of all earthquakes in the file
all_eq_dicts = all_eq_data['features']

# List to store all Magnitudes
mags, lons, lats, hover_texts = [], [], [], []
for eq_dict in all_eq_dicts:
  # Magnitude data is stored within the mag key that is under the properties section
  mag = eq_dict['properties']['mag']
  # Location data is stored under the key 'geometry' in a dictionary under this 
  # geometry key is a 'coordinates' key in a list with index 0 being lon, 1 being lat
  lon = eq_dict['geometry']['coordinates'][0]
  lat = eq_dict['geometry']['coordinates'][1]
  # 'title' will store name a summary of info on the earthquake
  title = eq_dict['properties']['title']
  # Add maginitude, lon, and lat to respective lists
  mags.append(mag)
  lons.append(lon)
  lats.append(lat)
  hover_texts.append(title)

# Map the earthquakes

# Create Scattergeo object inside list, which lets you overlay a scatter plot
# of geographic data on a map. Provide a list of longitudes and latitudes
# Simplest way to get data
# data = [Scattergeo(lon=lons, lat=lats)]
data = [{
    'type': 'scattergeo',
    'lon': lons,
    'lat': lats,
    # 'text' is what will label each marker/plot, we are including a summary of info here
    'text': hover_texts,
    # marker will change the look of the plots based on various options
    'marker': {
        # we are modifying the size of the 'marker' use a list comprehension
        # to get a size that will show up well
        'size': [3*mag for mag in mags],
        # 'color' tells plotly what values should be used to determine colorscale
        'color': mags,
        # 'colorscale' tells plotly what range of colors to use
        'colorscale': 'Viridis',
        # 'reversescale' will make the colorscale scale backwards
        'reversescale': True,
        # 'colorbar' affects the key of the scale on the side, we are titling it
        'colorbar': {'title': 'Magnitude'}
    },
}]


# Create Layout
graph_title = all_eq_data['metadata']['title']
my_layout = Layout(title=graph_title)

# fig is a dictionary that contians the data and layout
fig = {'data': data, 'layout': my_layout}
# Pass fig and filename to plot fucntion
offline.plot(fig, filename='global_earthquakes.html')

figs = go.Figure(fig)
figs.show()
```

Brightness of Global Fires Over 1 Day Period

```python
#@title Brightness of Global Fires Over 1 Day Period
import csv
from datetime import datetime

from plotly.graph_objs import Scattergeo, Layout
from plotly import offline

filename = 'drive/MyDrive/csv_file_data/world_fires_1_day.csv'
with open(filename) as f:
  reader = csv.reader(f)
  header_row = next(reader)

  dates, brightnesses, hover_texts = [], [], []
  lats, lons = [], []
  for row in reader:
    date = datetime.strptime(row[5], '%Y-%m-%d')
    brightness = float(row[2])
    label = f"{date.strftime('%m/%d/%y')} - {brightness}"

    dates.append(date)
    brightnesses.append(brightness)
    lats.append(row[0])
    lons.append(row[1])
    hover_texts = (label)

# Map the fires
data = [{
    'type': 'scattergeo',
    'lon': lons,
    'lat': lats,
    'text': hover_texts,
    'marker': {
        'size': [brightness/35 for brightness in brightnesses],
        'color': brightnesses,
        'colorscale': 'YlOrRd',
        'reversescale': False,
        'colorbar': {'title': 'Brightness'}
    },
}]

my_layout = Layout(title= 'Brightness of Fires Across the Globe in 1 Day')

fig = {'data': data, 'layout': my_layout}

offline.plot(fig, filename='world_fires_1_day.html')
```

```python
#@title Making a Bar Graph with Info on GitHub's Most Popular Repos
# requests is a package that lets you request info from a website
import requests

from plotly.graph_objs import Bar
import plotly.graph_objects as go
from plotly import offline

# Make an API call and store the response

# Store the url in a variable
url = 'https://api.github.com/search/repositories?q=language:python&sort=stars'
# Define headers to specifically use github's 3rd version of the API
headers = {'Accept': 'application/vnd.github.v3+json'}
# Use requests to make a call to the API, use get() and pass url and header
r = requests.get(url, headers=headers)
# r stores the response and it has an attribute 'status_code' that will be 200
# if succesful
print(f"Status code: {r.status_code}")

# Process results

# Store API response in a variable
# API returns info in a JSON format so we use json() to convert to python dict
response_dict = r.json()
repo_dicts = response_dict['items']
repo_links, stars, labels = [], [], []
# For loop to get the name and amount of stars in respective lists to graph
for repo_dict in repo_dicts:
  # Create a link with the name a of the repo and a link to it
  repo_name = repo_dict['name']
  repo_url = repo_dict['html_url']
  # Use HTML to make the name into a clickable link
  repo_link = f"<a href='{repo_url}'>{repo_name}</a>"
  repo_links.append(repo_link)

  stars.append(repo_dict['stargazers_count'])

  # Create label for each bar in the Bar graph
  owner = repo_dict['owner']['login']
  description = repo_dict['description']
  # Plotly lets you use HTML code so we make a line break with <br />
  label = f"{owner}<br />{description}"
  labels.append(label)


# Make visualization

# Set up the data specifying type and axes
data = [{
    'type': 'bar',
    'x': repo_links,
    'y': stars,
    'hovertext': labels,
    'marker': {
        'color': 'rgb(60, 100, 150)',
        'line': {'width': 1.5, 'color': 'rgb(25, 25, 25)'}
    },
    'opacity': 0.6,
}]
# We didn't need to import Layout Class since we made the layout ourselves with 
# a dictionary
my_layout = {
    'title': 'Most-Starred Python Projects on GitHub',
    'titlefont': {'size': 28},
    'xaxis': {
        'title': 'Repository',
        'titlefont': {'size': 24},
        'tickfont': {'size': 14},
        'tickangle': 45
    },
    'yaxis': {
        'title': 'Stars',
        'titlefont': {'size': 24},
        'tickfont': {'size': 14},
    },
}

fig = {'data': data, 'layout': my_layout}
offline.plot(fig, filename='python_repos.html')

figgo = go.Figure(fig)
figgo.show()
```

```python
#@title Display the posts with the most comments using an API
from operator import itemgetter

import requests 


# Make an API call, and store the response
url = 'https://hacker-news.firebaseio.com/v0/topstories.json'
r = requests.get(url)
print(f"Status code: {r.status_code}")

# Process information about each submission
# Convert into python list, the list features the id of the top posts
submission_ids = r.json()
# Create a list to hold a list of submission dictionaries
submission_dicts = []
for submission_id in submission_ids[:5]:
  # Make a seperate API call for each submission
  url = f"https://hacker-news.firebaseio.com/v0/item/{submission_id}.json"
  r = requests.get(url)
  print(f"id: {submission_id}\tstatus: {r.status_code}")
  response_dict = r.json()

  # Build a dictionary for each article

  submission_dict = {
      'title': response_dict['title'],
      'hn_link': f"http://news.ycombinator.com/item?id={submission_id}",
      'comments': response_dict['descendants'],
  }
  submission_dicts.append(submission_dict)

# Store each submission dictionary in a list
# ittemgetter('key') will pull the value associated with 'key'
submission_dicts = sorted(submission_dicts, key=itemgetter('comments'), reverse=True)

for submission_dict in submission_dicts:
  print(f"\nTitle: {submission_dict['title']}")
  print(f"Discussion link: {submission_dict['hn_link']}")
  print(f"Comments: {submission_dict['comments']}")
```
